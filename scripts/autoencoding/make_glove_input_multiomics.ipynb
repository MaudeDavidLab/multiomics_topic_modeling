```python
import pandas as pd
import numpy as np
import csv
```

```python
mtg = pd.read_csv("C:/Users/kitikomp/Documents/Lab/M3/data/mtg/asv_table_nooutliers.txt", sep = '\t')
metabol = pd.read_csv("C:/Users/kitikomp/Documents/Lab/M3/data/metabol/asv_table_nooutliers.txt", sep = '\t')
mtg = mtg.transpose()
metabol = metabol.transpose()
```

```python
print(mtg.shape)
print(metabol.shape)
```

```python
both = mtg.join(metabol, how = "inner")
both.shape
```

```python
both_norm = (both) / both.std(axis = 0)
```

```python
keep = []
for i in np.arange(both_norm.shape[1]):
    present = np.sum(both_norm.iloc[:, i] > 0)  > 0.1 * both_norm.shape[0]
    keep = keep + [present]
```

```python
both_norm = both_norm.loc[:, keep]
both_norm.shape
```

```python
feature_names = both_norm.columns.values
feature_names = np.array([i.replace(" ", "") for i in feature_names])
present = both_norm > 0
```

```python
len(feature_names)
```

```python
outfile = open("C:/Users/kitikomp/Documents/Lab/M3//results/glove/mtg_metabol_glove_input.txt", mode = "w")
writer = csv.writer(outfile, delimiter = "\t", quoting = csv.QUOTE_NONE, escapechar = '')
for i in np.arange(present.shape[0]):
    writer.writerow(feature_names[present.iloc[i, :]])
outfile.close()
```
